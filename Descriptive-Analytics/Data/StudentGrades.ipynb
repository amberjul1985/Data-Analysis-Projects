```python
import pandas as pd #installing pandas to work with data
df = pd.read_csv('Student_Grades.csv') #creating a dataframe using student grade data from CSV file
print(df.head()) #produce first 5 rows of table (including all columns)
print(df.describe())
```

```python
df_1 = df.drop(['Name', 'StudentID'], axis=1)
print(df_1)
skewness = df_1.skew()
kurtosis = df_1.kurtosis()

print(skewness)
print(kurtosis)
```

```python
dfdr = (df.dropna())
print(dfdr.describe())
```

```python
dfme = df.fillna(df.mean(numeric_only=True))
print(dfme.describe())
```

```python
dfmd = df.fillna(df.median(numeric_only=True))
print(dfmd.describe())
```

```python
dffi = df.ffill()
print(dffi.describe())
```

```python
dfba = df.bfill()
print(dfba.describe())
```
